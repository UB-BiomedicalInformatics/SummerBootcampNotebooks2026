```python
# KEYS AND FOREIGN KEYS
# LETS LOOK AT THE obs TABLE
sql_query = "DESC obs "
df = pd.read_sql_query(sql_query, conn)
df
# IN THE OPENMRS DATABASE, 'concept' DESCRIBES THE OBSERVAIONS/VISITS
# person_id AND concept_id ARE FOREIGN KEYS TO THE obs TABLE
```

```python
# JOINS
# person_id AND concept_id COLUMNS IN THE obs TABLE ARE FOREIGN KEYS 
# IF WE WANTED TO LOOK AT THE OBSERVATIONS FOR A PATIENT BASED ON person_id
# WE SHOULD JOIN THE obs, person_name, AND concept_name TABLES
sql_query = "select obs.person_id, pn.given_name, pn.middle_name, pn.family_name, con.name as concept_name, obs_datetime"
sql_query = sql_query + " FROM obs " 
sql_query = sql_query + " LEFT JOIN person_name pn ON obs.person_id = pn.person_id " 
sql_query = sql_query + " LEFT JOIN concept_name con ON obs.concept_id = con.concept_id" 
sql_query = sql_query + " WHERE obs.person_id in (50)" 
df = pd.read_sql_query(sql_query, conn)
df
```

```python
# MORE JOINS
# THE name COLUMN OF concept_name TABLE DOES NOT PROVIDE ENOUGH INFORMATION, IN CASE WE WANTED A MORE DETAILED DESCRIPTON OF THE VISIT 
# WE CAN INCLUDE ANOTHER TABLE concept_description. THE concept_id COLUMN IS A FOEIGN KEY IN THE concept_description TABLE.  
sql_query = "select obs.person_id, pn.given_name, pn.middle_name, pn.family_name, con.name as concept_name, cd.description as concept_desc, obs_datetime"
sql_query = sql_query + " FROM obs" 
sql_query = sql_query + " LEFT JOIN person_name pn ON obs.person_id = pn.person_id" 
sql_query = sql_query + " LEFT JOIN concept_name con ON obs.concept_id = con.concept_id" 
sql_query = sql_query + " LEFT JOIN concept_description cd on obs.concept_id = cd.concept_id"
sql_query = sql_query + " WHERE obs.person_id in (50)" 
df = pd.read_sql_query(sql_query, conn)
# THIS LINE SAVES THE RESULT SET AS A FILE SINCE JUPYTER NOTEBOOK DISPLAY IS TRUNCATED. 
df.to_csv('concept_description_example.csv', encoding='utf-8', index=False, header=True, mode='w')
# YOU SHOULD BE ABLE TO SEE THE FILE, concept_description_example.csv, IN THE LEFT HAND PANE ONCE YOU RUN THIS CELL
# IF YOU HAVE THE FILE OPEN AND RUN THE CELL YOU MAY GET A PERMISSION DENIED ERROR, SO CLOSE THE FILE BEFORE YOU (RE)RUN THE CELL.
df
```

```python
# GROUPING AND AGGREGATION
# FIND THE COUNT OF ALL OBSERVATION/VISIT TYPES FOR PATIENT/PERSON NO. 18 AND SORT THEM IN DESCENDING ORDER OF FREQUENCY
sql_query = "SELECT person_id, con.name,cd.description, COUNT(*)"
sql_query = sql_query + " FROM obs " 
sql_query = sql_query + " LEFT JOIN concept_name con ON obs.concept_id = con.concept_id" 
sql_query = sql_query + " LEFT JOIN concept_description cd ON obs.concept_id = cd.concept_id" 
sql_query = sql_query + " WHERE person_id IN (18)" 
sql_query = sql_query + " GROUP BY  person_id, con.name, cd.description" 
sql_query = sql_query + " ORDER BY COUNT(*) DESC" 
df = pd.read_sql_query(sql_query, conn)
df 
```

```python
# MORE GROUP BY AND AGGREGATION
# FIND ALL PATIENTS WHO HAVE HAD A HEART RATE MEASUREMENT IN OBSERVATIONS TABLE AND SORT THEM IN DESCENDING ORDER 
# ACCORDING TO THE TOTAL OBSERVATIONS
sql_query = "SELECT person_id,count(*)"
sql_query = sql_query + " FROM obs " 
sql_query = sql_query + " LEFT JOIN concept_name con ON obs.concept_id = con.concept_id" 
sql_query = sql_query + " where con.name in ('HEART RATE') AND obs.value_numeric is not null" 
sql_query = sql_query + " GROUP BY  person_id" 
sql_query = sql_query + " ORDER BY COUNT(*) DESC" 
df = pd.read_sql_query(sql_query, conn)
df 
```

```python
# HAVING CLAUSE
# FIND MOST FREQUENT VISIT TYPES BY PATIENT THOSE PATIENTS GREATER THAN 60 YRS OF AGE
sql_query = "SELECT obs.person_id, TIMESTAMPDIFF(YEAR, p.birthdate, CURDATE()) AS age, con.name as concept_name, count(*)"
sql_query = sql_query + " FROM obs " 
sql_query = sql_query + " LEFT JOIN concept_name con ON obs.concept_id = con.concept_id" 
sql_query = sql_query + " LEFT JOIN person p ON obs.person_id = p.person_id" 
sql_query = sql_query + " WHERE TIMESTAMPDIFF(YEAR, p.birthdate, CURDATE()) >= 60" 
sql_query = sql_query + " GROUP BY  obs.person_id, con.name HAVING age >= 60" 
sql_query = sql_query + " ORDER BY person_id, COUNT(*) DESC" 
df = pd.read_sql_query(sql_query, conn)
df 
```

```python
# QUANTIFIERS
# FIND ALL OBSERVATION TYPES FOR PATIENTS OVER THE AGE OF 75
sql_query = "SELECT DISTINCT con.name as concept_name"
sql_query = sql_query + " FROM obs " 
sql_query = sql_query + " LEFT JOIN concept_name con ON obs.concept_id = con.concept_id" 
sql_query = sql_query + " LEFT JOIN person p ON obs.person_id = p.person_id" 
sql_query = sql_query + " WHERE TIMESTAMPDIFF(YEAR, p.birthdate, CURDATE()) >= 80" 
sql_query = sql_query + " ORDER BY con.name" 
df = pd.read_sql_query(sql_query, conn)
df 
```

```python
# NESTED QUERIES - LINES 7-10 ARE PART OF THE NESTED QUERY   
# FIND ALL PATIENTS WHO HAVE HAD A HEIGHT MEASUREMEMENT AND A WEIGHT MEASUREMENT
sql_query = "SELECT DISTINCT person_id"
sql_query = sql_query + " FROM obs " 
sql_query = sql_query + " LEFT JOIN concept_name con ON obs.concept_id = con.concept_id" 
sql_query = sql_query + " WHERE con.name IN ('WT', 'WEIGHT (KG)') AND VALUE_NUMERIC IS NOT NULL AND person_id IN (" 
sql_query = sql_query + "   SELECT DISTINCT person_id" 
sql_query = sql_query + "   FROM obs" 
sql_query = sql_query + "   LEFT JOIN concept_name con ON obs.concept_id = con.concept_id" 
sql_query = sql_query + "   WHERE con.name IN ('HT', 'HEIGHT (CM)') AND VALUE_NUMERIC IS NOT NULL" 
sql_query = sql_query + "   ) " 
sql_query = sql_query + " ORDER BY obs.person_id " 
df = pd.read_sql_query(sql_query, conn)
df 
```
