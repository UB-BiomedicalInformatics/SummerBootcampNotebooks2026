## Tree classifiers exercise
We have covered the basics of constructing decision trees and random forests. In this exercise, you will apply those skills to Autistic Spectrum Disorder dataset. This dataset contains some basic demographic information about patients and the patient's answers (scored as a `0` or `1`) to a series of screening questions. 

The data is contained in the file named `autism_adolescent.csv` located in the same directory as this file. The data elements are defined as follows:
* `A1_Score`: The answer to question 1; values: 0, 1
* `A2_Score`: The answer to question 2; values: 0, 1
* `A3_Score`: The answer to question 3; values: 0, 1
* `A4_Score`: The answer to question 4; values: 0, 1
* `A5_Score`: The answer to question 5; values: 0, 1
* `A6_Score`: The answer to question 6; values: 0, 1
* `A7_Score`: The answer to question 7; values: 0, 1
* `A8_Score`: The answer to question 8; values: 0, 1
* `A9_Score`: The answer to question 9; values: 0, 1
* `A10_Score`: The answer to question 10; values: 0, 1
* `jaundice`: Did the patient have jaunice; values: 0 (no), 1 (yes)
* `gender`: The patient's gender; values 0 (male), 1 (female)
* `autism`: Does the patient have autism; values: 0 (no), 1 (yes)
