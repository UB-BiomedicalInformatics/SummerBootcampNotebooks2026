```r
# Let's learn some decision structure

student1 <- "Anjali"

if (student1 == "Anjali") { ## "==" tests whether left element is equal to right element
    print("Anjali is a student in ENABLE 2019.")
}
```

```r
# What if there are more than one decision? 

student1 <- "Zipporah"

if (student1 == "Zipporah") { ## "==" tests whether left element is equal to right element
    print("Zipporah is originally from Indiana.")
} else {
    print("This student is not originally from Indiana")
}

student1 <- "Anjali"

if (student1 == "Zipporah") { ## "==" tests whether left element is equal to right element
    print("Zipporah is originally from Indiana.")
} else {
    print("This student is not originally from Indiana")
}
```

```r
# What if there are more than one condition? 
# For instance, you might want to test whether registered students are all from Bennett College

student1 <- "Shaneer"
student2 <- "Alexis"

if (student1 == "Shaneer" & student2 == "Alexis" ) { ## "&" is logical AND operator which gives a output TRUE if both the elements are TRUE.
    print("Both students are from Bennett College.")
} else {
    print("One or both students are not from Bennett College.")
}


student1 <- "Shaneer"
student2 <- "Ayesha"

if (student1 == "Shaneer" & student2 == "Alexis" ) { ## "&" is logical AND operator which gives a output TRUE if both the elements are TRUE.
    print("Both students are from Bennett College.")
} else {
    print("One or both students are not from Bennett College.")
}

```

```r
# Logical OR operator ("|") tests whether one of elements is TRUE

student1 <- "Shaneer"
student2 <- "Alexis"

if (student1 == "Shaneer" | student2 == "Alexis" ) { ## "&" is logical AND operator which gives a output TRUE if both the elements are TRUE.
    print("At least, one of students are from Bennett College.")
} else {
    print("None of them are from Bennett College.")
}

student1 <- "Shaneer"
student2 <- "Ayesha"

if (student1 == "Shaneer" | student2 == "Alexis" ) { ## "&" is logical AND operator which gives a output TRUE if both the elements are TRUE.
    print("At least, one of students are from Bennett College.")
} else {
    print("None of them are from Bennett College.")
}

student1 <- "Chanell"
student2 <- "Ayesha"

if (student1 == "Shaneer" | student2 == "Alexis" ) { ## "&" is logical AND operator which gives a output TRUE if both the elements are TRUE.
    print("At least, one of students are from Bennett College.")
} else {
    print("None of them are from Bennett College.")
}
```

```r
# Let's learn about "function."
# A function is a set of statements organized together to perform a specific task. 
# R has a large number of in-built functions and the user can create their own functions.

# let's look at in-built functions first. 

# You can pass parameter(s) to the function. You can consider parameter(s) as input values. 
# Parameters can be put within parentheses. In the following example, 10 is the parameter.
# 2.302... is returend value (output value) from the function.

log(10)

# Another function example
log10(10)
```

```r
# You can create your own functions.
# For instance, statements in the for loop in the above example can be converted as a function.
# If you are tasks that are repeated many times, it is generally a good idea to create functions.
# Functions can make your program more modular and maneageable.

decide_tell_Bennett <- function (value) {
    if (value == "Shaneer" | value == "Alexis" ) {
        # cat is a function that concatenate strings and pring them
        # "\n" stands for new line. It's like "enter" in Windows and "return" in Mac
        cat(value , " is from Bennett College.", "\n")
    } else {
        cat(value , " is not from Bennett College.", "\n")
    }
}

# Let's simply print each element in the list
for (val in enable_2019) {
    decide_tell_Bennett(val)     
}
```

```r
# In the previous version of decide_tell_Bennett function,
# strings were printed within function. Instead, you can return value(s) to the place where the function is called.
# Then, print the values within for loop.


decide_tell_Bennett <- function (value) {
    if (value == "Shaneer" | value == "Alexis" ) {
        # cat is a function that concatenate strings and pring them
        # "\n" stands for new line. It's like "enter" in Windows and "return" in Mac
        return (cat(value , " is from Bennett College.", "\n"))
    } else {
        return (cat(value , " is not from Bennett College.", "\n"))
    }
}

# Let's simply print each element in the list
for (val in enable_2019) {
    decide_tell_Bennett(val)
}
```
