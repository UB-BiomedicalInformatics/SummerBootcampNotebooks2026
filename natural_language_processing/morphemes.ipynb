## Morphemes

Morphemes are the building blocks of meaning that words are composed of.  Linguists generally divide morphemes into two types: inflectional and derivational.  Derivational morphemes are those that change the meaning of a word to create a new word.   For example, if you add the prefix “un-” to a word like “tie” you get a new word “untie” that means essentially the opposite.  Inflectional morphemes, on the other hand, are morphemes that don’t change the meaning of the word, but are required by syntax to provide information such as tense or number.   If we add the past tense suffix “-ed” to a verb like “retreat” we get “retreated” which still has the original meaning of “retreat”.

In information extraction, we are typically most interested in inflectional morphology.   The process of removing the inflectional morphemes from a word is known as “stemming”, with the “stem” of a word being the core of the word with its inflectional morphemes removed.   The “root” of a word is essentially the same thing with the derivational morphemes also removed.

Information extraction applications often stem the words in a sentence in order to get at the meanings of the words without having to account for the additional meanings of the inflectional morphemes. For example, if we have a biomedical informatics application where we want to look up all the patients who have a tumor, we may also want to include patients who have multiple tumors.  In this case, the plural “s” at the end of “tumors” needs to be accounted for.

How would you write an algorithm to accomplish this?   

For the most part, inflectional morphology is regular, that is, it works the same way for each word in the language. However, there are exceptions.  The plural of “mouse” is “mice”.  The past tense of “sink” is “sank”. 

Even the regular rules are more complicated than you might at first suspect.   For example, we can usually simply add an “-ed” to the end of a word to get a past-participle.   However, if the word ends in a consonant and then a “y”, we change the “y” into an “i” before adding the “-ed”. If a verb’s final syllable has a short vowel like the “a” sound in “scam”, then we must double the final consonant before adding “-ed”.  In stemming a verb like “scammed”, we would have to know that we need to remove the “-ed” but also the final “m” to get back to the stem of the word.

The most famous algorithm for doing stemming is the Porter Stemmer. The Porter Stemmer attempts to take into account all of these variations, but does not take into account irregular inflections.

Try using the NLTK implementation of the Porter Stemmer with regular and irregular words
