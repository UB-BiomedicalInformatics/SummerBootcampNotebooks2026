## Information Extraction
The expression “Natural Language Processing” or “NLP” is used to cover a wide variety of technologies involving the interaction between natural human language (i.e. English, Italian, Mandarin etc.) and computers.   These technologies can include everything from the automatic speech recognition done by Dragon NaturallySpeaking to machine translation as done by Google Translate.  In this course we will cover the much narrower topic of information extraction from machine readable text, sometimes called “text exploitation”.   The general idea of information extraction is much of the knowledge available in the world is in the “unstructured” form of written communication and in order to interpret this information we need a computer to extract the relevant parts of it out so that it can be used in an application.   Examples of this might include scanning a medical record to automatically see what procedures were performed for the sake of billing, or analyzing a large number of such records in the course of doing a retrospective study on a particular drug.

In order to understand the technologies of NLP, we need to have a general understanding of linguistics in order to see how the technologies map onto various aspects of natural language.  The field of linguistics is very broad and it covers many topics that we don’t need to understand in order to do information extraction, such as knowing how speech sounds are produced or recognized.  However, most sub-areas of linguistics serve an important function in many information extraction applications.
We can generally think of linguistics as being organized hierarchically with language being built up from speech sounds - phonemes, to larger units of meaning - morphemes,  to words, to phrases and sentences, to groups of sentences in a narrative or discourse, with the general idea of meaning being composed of how all these things work together.   Each of these levels has a corresponding linguistic subfield, phonology, morphology, lexicology, syntax, discourse analysis and semantics being the study of meaning.   Each of these areas also has some amount of corresponding coding techniques for handling them.  In this lesson we will cover each of these topics with respect to understanding their function in the context of getting information out of natural language text so that it can be used in a further application.

Even within each of these areas there are broadly speaking two different approaches to handling them: symbolic versus statistical.  The symbolic approaches to NLP are those that use deterministic hand-designed algorithms to get their results.  The statistical approaches are those that use machine learning techniques.  There is not a hard and fast rule about when each technique is appropriate and almost every fully developed NLP application uses some combination of statistical and symbolic techniques.  For example, a typical NLP system might use symbolic techniques to break up the text into sentences and words, but then use a statistical algorithm to combine those words into useable phrases.
