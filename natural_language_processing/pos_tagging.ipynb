## POS Tagging

Many information extraction applications involve a step of linguistic analysis in which words are assigned a **part-of-speech**, called POS Tagging.   A part-of-speech is a designation of the syntactic role that a word can play in a sentence. You remember learning in grade school that nouns represent people, places and things, whereas verbs are “action words”, and adjectives are describing words.   

In modern formal linguistics there is often little difference between the theoretical notion of part-of-speech and a set of selectional restrictions that a word might have in a theoretical lexicon. For the purposes of NLP, the parts have speech have largely been restricted to those that have been annotated in a large corpus of English data called the Penn Tree Bank. Linguists at the University of Pennsylvania have painstakingly annotated this corpus for the purpose of designing and testing tools and algorithms that could reproduce or use the results. The parts-of-speech in the Penn Tree Bank are as follows:

* CC Coordinating Conjunction  ('and','but', etc.)
* CD Cardinal Number ("four")
* DT Determiner ("a","the", "this" as in "This sandwich is great.")
* EX Existential There ("There are a lot of parts-of-speech.")
* FW Foreign Word (All foreign words are given this tag regardless of their part-of-speech)
* IN Preposition ("in","on","of" etc.)
* JJ Adjective ("blue","happy")
* JJR Adjective, comparative ("bluer","happier")
* JJS Adjective, superlative ("bluest","happiest")
* LS List item marker
* MD Modal ("can","could","will","would","shall","should","may","might")
* NN Noun, singular or mass ("cat", "cotton")
* NNS Noun, plural ("cats")
* NNP Proper noun, singular ("William Shakespeare","NASA")
* NNPS Proper noun, plural ("Williams" as in "There are three Williams in my class.")
* PDT Predeterminer ("both")
* POS Possessive ending ("'s")
* PRP Personal pronoun ("he","him","she","her","it","they","them")
* PRPS Possessive pronoun ("his","hers","its","their")
* RB Adverb ("quickly")
* RBR Adverb, comparative ("less" or "more" as in "He ran less quickly.")
* RBS Adverb, superlative ("least" or "most" as in "He ran most quickly.")
* RP Particle (these look like prepositions but they are really part of the verb. "off" in "I turned the light off.")
* SYM Symbol ("=" etc.)
* TO to (literally all instances of "to")
* UH Interjection ("hmm")
* VB Verb, base form ("run" as in "I like to run.")
* VBD Verb, past tense ("ran")
* VBG Verb, gerund or present participle ("running" as in "I am running." or "I see running people.")
* VBN Verb, past participle ("run" as in "I have run a lot.")
* VBP Verb, non-3rd person singular present ("run" in "I run a lot.")
* VBZ Verb, 3rd person singular present ("runs" in "He runs a lot.")
* WDT Wh-determiner ("which" in "which candidate do you prefer.")
* WP Wh-pronoun ("who","what","where","why","how")

Notice there are a few simplifications that the Penn Tree Bank makes in order to facilitate annotation.  For example, the part-of-speech TO is used for every instance of the word “to”, whether it is being used as a locative preposition or an infinitive verb marker.

A naive algorithm for getting the parts-of-speech would probably start with simply listing all the words with a single part-of-speech. This would ignore ambiguous words like “can” which, depending on context, may be a noun, a verb or a modal. It would also ignore the fact that English often allows for zero-morphology derivations where a word that typically has one part of speech may be used for another part of speech, such as the case of “verbing”.

Part-of-speech tagging is an area of NLP that is typically handled using a statistical method like hidden Markov models.
