## Chunking

Linguistic syntax is usually conceived of in terms of generative grammars that use recursive rules. For example, a rule might be that a Noun Phrase can consist of a DT NN and a Prepositional Phrase, where DT and NN are parts of speech from the previous section and a Prepositional Phrase can be rewritten as a IN followed by a Noun Phrase.   

Linguists test these rules to ensure that they confirm our intuitive notion of constituency: the idea that each rule should account for something that we actually consider to be a natural part of a sentence. In a sentence like, “The queen of England is rich”, we do not consider “The queen” to be a constituent by itself because then we would have to argue that “of England” is somehow participating in the sentence separately from “the queen”, which it does not.

However, in NLP elements such as “The queen” prove to be useful for the purposes of getting a shallow interpretation of the text. Such elements are called chunks. In the case of Noun chunks, they are basically a noun phrase up to the head noun and not including any subsequent modifiers such as prepositional phrases.   The idea is that if we want to know who is rich, then it will suffice to say that it is the queen, and shortcuts having to make a decision about whether “of England” is a modifier of “the queen” or not.  The process of producing chunks is called chunking.  An NLP system typically might produce noun chunks and verb chunks.

Chunking algorithms are usually statistical, often using hidden Markov models to determine the starting point and end point of chunks based on annotated data. Basic chunking in NLTK, however, is done using rules.

In the following example, the rule for creating chunks is that chunks consist of a DT followed by an NN.


```python
from nltk.chunk import *
from nltk.chunk.util import *
from nltk.chunk.regexp import *
from nltk import word_tokenize, pos_tag, Tree

pos_tagged_sentence = pos_tag(word_tokenize("This is a test sentence."))

tag_pattern = "<DT><NN>"
rule = ChunkRule(tag_pattern,"Determiner Noun")
chunk_parser = RegexpChunkParser([rule], chunk_label="NounPhrase")
Tree.pretty_print(chunk_parser.parse(pos_tagged_sentence))
```

The tag_pattern follows the rules of python regular expressions, so if you want to add an optional adjective phrase it would look like

```python
tag_pattern = "<DT><JJ>?<NN>"
rule = ChunkRule(tag_pattern,"Determiner Noun")
chunk_parser = RegexpChunkParser([rule], chunk_label="NounPhrase")
Tree.pretty_print(chunk_parser.parse(pos_tagged_sentence))
```

Or if you needed multiple adjectives:

```python
tag_pattern = "<DT><JJ>*<NN>"
rule = ChunkRule(tag_pattern,"Determiner Noun")
chunk_parser = RegexpChunkParser([rule], chunk_label="NounPhrase")
Tree.pretty_print(chunk_parser.parse(pos_tagged_sentence))
```

Or for 1 or more adjectives:

```python
tag_pattern = "<DT><JJ>+<NN>"
rule = ChunkRule(tag_pattern,"Determiner Noun")
chunk_parser = RegexpChunkParser([rule], chunk_label="NounPhrase")
Tree.pretty_print(chunk_parser.parse(pos_tagged_sentence))
```
