## Tokenization

 The process of breaking down some text into its component words is called tokenization.  A "token" is an instance of a word as opposed to the abstract version of the word that might exist in the dictionary or as a part of human grammar.  In a sentence like "I am here and I am happy." there are only five words because "I" and "am" is repeated but there are seven separate tokens.
 
Take the same example and attempt to tokenize it using String.split().

```python
text = "Mr. Smith's mother lives at 1234 Maple Rd. His brother. Dr. Smith says, 'I love my mother. She is one of my favorite people.'"
```

The NLTK method of doing this would look as follow:

```python
import nltk

from nltk import word_tokenize

word_tokenize(text)
```

What do you notice is different about this output than the naive sollution?  Did you notice that "'s" is a token?
