## Sentence Breaking

NLP applications consist of many low level task such as determining which parts of a text are individual sentences for doing analysis on.  If you don’t think about it too closely these tasks might seem trivial.  One simply breaks each sentence apart based on the location of periods, question marks or exclamation marks.   However, take a look at the following text:


```python
text = "Mr. Smith's mother lives at 1234 Maple Rd. His brother. Dr. Smith says, 'I love my mother. She is one of my favorite people.'"
```

Use the String.split() method on the text to see what the results are.

It is possible, depending on your application that you do not care whether “Mr” by itself counts as a full sentence, but many applications do.   For example, if you want to identify all the people that occur in a sentence, it would be useful to be able to identify all those with the word “Mr.” in front of them.

Suppose that we had a list of all the abbreviations in our texts that end in periods.  How would you do this differently?

```python
abbreviations = ["Mr.","Rd.","Dr."]
```

Did you account for the fact that sentences begin with a capital letter?

Does the solution work for the following text?

```python
text2 = "I live at 123 Maple St. My brother lives on 123 Elm Dr. His name is John."
```

In the ideal world there would also be some recognition that often there are multiple spaces after a period but not necessarily so after an abbreviation.

Try the same texts using NLTK:

```python
from nltk.tokenize import sent_tokenize

sent_tokenize(text)

from nltk.tokenize import punkt

tokenizer = punkt.PunktSentenceTokenizer()

tokenizer.tokenize(text)
```

The default tokenizer is using rules similar to those that you may have written above.   It is a symbolic algorithm and requires no training and is veryfast.   The PunktSentenceTokenizer is a statistical algorithm, is slightly slower, and may require training if your text has different types of sentences than are already modeled.
