## Named Entities

Chunking is often used as a means of extracting named entities.   These are usually proper nouns that are categorized into useful elements for the application.  For example if you are building an application that wants to identify which patients are taking which drug, a named-entity identifier that gets the semantic concepts of persons and drugs would be useful.

The default named-entity recognizer in NLTK recognizes a small ontology from the NIST Automatic Content Extraction competition, which includes PERSONS, ORGANIZATIONS and GPE, which are geopolitical entities. 



```python
import nltk

from nltk.corpus import conll2000
from nltk.chunk import conlltags2tree, tree2conlltags
from nltk.chunk import ne_chunk
from nltk import pos_tag

s = "Harry Potter went to the Hogwarts Academy For Wizardry."
t = nltk.word_tokenize(s)
tags = nltk.pos_tag(t)

nes = ne_chunk(tags)

print(nes)
```

What NLTK actually returns from this is a tree where some of the nodes on the tree are labeled with an entity type, so we need a recursive function to read it:

```python
def getNodes(parent):
    for node in parent:
        if type(node) is nltk.Tree:
            print("Label:",node.label())
            print("Leaves:", node.leaves())
            getNodes(node)
            
getNodes(nes)
```

Once named entities are identified it is possible to establish relations between entities.

Often it is is actually much simpler to avoid complicated syntax to get the kinds of results that an information extraction algorithm requires. If you have a sentence that contains a patient name and a drug for example, you will know that that patient is in some way involved with that drug. If the application is to find all the people who are likely to benefit from a follow-up to prevent opioid addiction, then this may be sufficient for finding these patients with a human to look at the data later.

The details of syntactic theory are far beyond the scope of this module. Modern syntax was developed in the 1950s by Noam Chomsky who accounted for the meanings of sentences by proposing that sentences have an "underlying form" that puts all of the the semantic participants in a place where they can be directly interpreted; e.g., the subject of the verb "run" is the "runner".  He then accounted for variations in syntax, such as the passive construction, by proposing that "transformations" on the sentence word order would then be performed to get the "surface form" of the sentence that is actually produced.

More recent syntactic theories have largely done away with "transformations", but retained much of Chomsky's ideas about how sentences are recursively formed. Dependency Grammar on the other hand starts not from the generative approach, but more on the question of how to interpret the relations between semantic elements of the sentence, which makes it an especially valuable theory for the purposes of information extraction. Dependency grammar is the basis of many relation extraction systems such as those that use Stanford Core NLP.

While NLTK does have some facilities for getting relations, it does not have a native dependency grammar library. It can use the Stanford Dependency Parser if you are running NLTK in Jython, but this requires a lot of setup.

The native NLTK relationships are highly configurable depending on what kinds of relations you wish to extract.  By default it tries to establish relation between named entities without really assigning any semantics to those relations.  In the following example, you can see that NLTK proposes are relation between Harry, Potter and Hogwarts Academy For Wizardry, possibly with the "went to the" relationship since it doesn't actually have a Tree for that. 

```python
 nltk.sem.relextract.tree2semi_rel(nes)
```
