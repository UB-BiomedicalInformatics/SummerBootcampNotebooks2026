We will focus on a few examples here: creating substrings via "slicing", index(), strip(), lower(), upper(), and len().

Slicing works by using square brackets [ ] surrounding a start position, end position, and a "step", all separated by colons. These brackets are placed after the string or variable representing the string. It is important to note that these start and end positions utilize **zero based indexing**, which means that the first character in the string is at position 0, the second is at position 1, the third is at position 2, etc. The range also slices **before** the endpoint, meaning that an end point of 25 would not include z, despite the letter z being at index 25 due to the zero based indexing. This can be seen in the code below.

The step determines if any characters should be skipped. For example, utilizing a step of 2 dictates that only every other character will be included.

Please note that if you do not provide the information, the start defaults to the first character, the end defaults to after the last character, and the step defaults to 1.

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
print(alphabet[25])
print(alphabet[0:26])
print(alphabet[1:26])
print(alphabet[1:25])
```

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
newstring = alphabet[1:20:2]
print(newstring)
```

The example above creates a string containing the alphabet and sets it to the variable titled alphabet. We then slice it using [1:20:2], which takes the second character through the 20th character, excluding every other one.

You can also reverse the order of the characters in the string by using a negative sign in your step.

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
newstring = alphabet[::-1]
print(newstring)
```

**Loops** - A loop is a sequence of instructions that is repeated until a certain condition is reached. We will be discussing **while loops** and **for loops**.

An example of a while loop is listed below. We first set our "count" variable to 0 and then begin our while loop. As specified in the parentheses, the loop will continue iterating until our count value is no longer less than 4.

Since 0 < 4, we begin our first iteration. The set of instructions for the loop is indented. The first step of the loop is to add 1 to the count variable. The loop then prints "Hello world". With the set of instructions complete, we go back to the beginning of the loop.

Since 1 < 4, we reenter the loop and repeat the steps. This continues happneing until our count reaches 4 ( 4 < 4 is FALSE ), in which case we do not reenter the loop and the process is complete.

```python
count = 0
while(count < 4):
    count += 1
    print("Hello world!")
```

**List Comprehension and Data Structures:**

Now that you have begun to get an understanding of the basics in python, we can take it a step further and familiarize ourselves with a few functions and commands to easily modify lists for our purposes.

To begin with, we will look at **list comprehension**. This allows us to generate a new list from an existing one, leaving the original unmodified. This is often done using a single line of code and allows us to bypass using a for loop every time we want to modify a list.

The syntax for this involves defining what you want to print **for** each iteration through a list, and you can specify to do it only **if** some condition is met.

You can find an example below that prints each string from a list of DNA strings, only if the sequence "TATATAT" can be found in said string of DNA.

```python
DNA_Sequences = ["ACCAACTCCAGCTATATATCCTCGCCGGGAAGACTTCATCCCAG", "CCGCAGCAGTTCTCGAACTGGCCTCCTTGA", "TCTGCAACCTGGATCTGGGGGACTGCGGGCCAGGCGCGGCGTG", "CCTGCCCTCCTCTCCTCCCCAAGCCTGACAGCCCGCGAGCTGCCATATATATAGCAGGGCGCAGCCATGGGAAGAGGA", "GGAGGGCTAGGGAGCGGCGGCGGCGGCGGCAGCGGCAGCAGCAGCTGCTACAGTGGCGGTGGCGGCGGCGGCTGCTGCTGAG"]

TATA_Sequences = [x for x in DNA_Sequences if "TATATAT" in x]

print(TATA_Sequences)
```

**Computational Theory: Recursion** - Understanding recursion can help when attempting to solve more complicated computational problems. Recursion in python allows a function to call upon itself to help simplify a code. A good example of this comes when attempting to define a function that calculates factorials. We define functions using def. Following def, we add a function name and any required arguments in the parentheses. We then define the function below this line. Defining a function like this is useful because it allows us to call upon that function again later in the code. It is also necessary for recursion. See below for an example:

```python
def factorial(x):
    if x == 1:
        return 1
    else:
        return (x * factorial(x-1))

print(factorial(4))
```

To calculate 4 factorial, we would use the following calculation:

4 * 3 * 2 * 1 = 24

We are able to solve this problem by defining a function that calls upon itself.

As we first feed the number 4 into the function, we do not satisfy the "if" conditional (x does not equal 1). Thus, we move onto else and return:

4 * factorial(4-1)

Which would then feed back into the function and result in:

4 * 3 * factorial(3-1)

Until our x value is equal to 1 and the function no longer feeds into itself, resulting in our desired:

4 * 3 * 2 * 1

This sort of thinking can be extremely useful when coding in python.
