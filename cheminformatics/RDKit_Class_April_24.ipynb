## Molecular Fingerprinting

We can now begin to run analyses on these molecules. One way to do this is through molecular fingerprint comparisons. A molecular fingerprint is a way of depicting a molecular structure through bit strings, essentially breaking the molecule down into different fragments. For our purposes, we are going to generate Morgan fingerprints as bit vectors. In order to do this, you need to set a specific radius. The radius informs the program how large to make the bits. See the image below:

This image shows what happens when you set different radii, making the circles expand as it increases (radius of 1, radius of 2, etc.). The program creates a fingerprint by going to each non-hydrogen element in the molecule and observing x number of bonds away from that center, where x is your radius. Each of these structures are documented in a binary format as seen above.

For our fingerprints, we will also specify that we want to utilize features (useFeatures=True). Simply put, this makes the fingerprinting search/account for different functional groups (ketones, alcohols, etc.). This often increases the accuracy of comparisons. 

We will start by setting the Morgan Fingerprint for caffeine to the variable caf_fp. We will set the radius to 2 and specify that we will use features for our fingerprinting. See below:

```python
caf_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(caf, radius=2, useFeatures=True)
```

We then repeat this process for each of the other molecules.

```python
dop_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(dop, radius=2, useFeatures=True)
mor_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(mor, radius=2, useFeatures=True)
rem_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(rem, radius=2, useFeatures=True)
ser_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(ser, radius=2, useFeatures=True)
reg_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(reg, radius=2, useFeatures=True)
ace_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(ace, radius=2, useFeatures=True)
lev_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(lev, radius=2, useFeatures=True)
sor_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(sor, radius=2, useFeatures=True)
mys_fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(mys, radius=2, useFeatures=True)
```

It may also be helpful to set all the fingerprints into a single set that we can call upon later.

```python
fingerprints = [caf_fp,dop_fp,mor_fp,rem_fp,ser_fp,reg_fp,ace_fp,lev_fp,sor_fp,mys_fp]
```

## Similarity Metrics

We can now compare these fingerprints using similarity metrics, of which there are many. See the link below for a few examples:

https://chem.libretexts.org/Courses/Intercollegiate_Courses/Cheminformatics_OLCC_(2019)/6%3A_Molecular_Similarity/6.2%3A_Similarity_Coefficients

We will be focusing on calculating Tanimoto and Dice coefficients. Both of these metrics are zero to one measurements, with zero being completely distinct molecules and one being identical molecules. Both take into account the bits that the two molecules have in common and compare it to the ones that are distinct. You can see how both formulas are constructed below.

Once our fingerprinting is complete, it is easy to calculate similarity metrics with RDKit. For example, the function DataStructs.TanimotoSimilarity(x,y) can be used to calculate the Tanimoto score between two fingerprints, x and y. Similarly DataStructs.DiceSimilarity(x,y) can be used to calculate Dice coefficients. See below for a comparison between the fingerprints for caffeine and morphine. 

```python
DataStructs.TanimotoSimilarity(caf_fp,mor_fp)
```

```python
DataStructs.DiceSimilarity(caf_fp,mor_fp)
```

Notice how the two calculations give different coefficients. Due to the differences in their formula, Tanimoto tends to score more harshly when there are major differences between the molecules. With that being said, the Tanimoto coefficient is 0.05 and the Dice coefficient is 0.095, both of which are very low scores. Does this make sense? Let's review these two molecules:

```python
caf
```

```python
mor
```

Clearly, there are some structural differences which are reflective of their functional differences. We can also use simple for loops to calculate multiple similarity coefficients at once. The loop below will compare the fingerprint of morphine to every other fingerprint in our set. 

```python
for x in fingerprints:
    print(DataStructs.TanimotoSimilarity(mor_fp,x))
```

Most of our similarity metrics here are fairly low, except for two instances. We received a perfect score for the third molecule in the set and the final molecule in the set.

fingerprints = \[caf_fp,dop_fp,**mor_fp**,rem_fp,ser_fp,reg_fp,ace_fp,lev_fp,sor_fp,**mys_fp**\]

Keep in mind that 3D molecular structures can be downloaded directly from PubChem and Drug Bank.

Additional options for 3D visualization in Jupyter Notebook can be accessed with the right python codes. Here is an example of a function that we can define called drawit:

```python
def drawit(m, p=None, confId=-1):
    mb = Chem.MolToMolBlock(m, confId=confId)
    if p is None:
        p = py3Dmol.view(width=400, height=400)
    p.removeAllModels()
    p.addModel(mb, 'sdf')
    p.setStyle({'stick': {}})
    p.setBackgroundColor('0xeeeeee')
    p.zoomTo()
    return p.show()
```

By setting specific instructions for this drawit function, we are able to alter how the molecule is displayed within Jupyter Notebook. See the new display of remdesivir below.

```python
drawit(rem)
```

Note that this .mol file did not originally have hydrogen atoms incorporated into it and we never added them (recall the warning message from before). If you have a .mol file does, the following argument can be added when you initially load in the .mol to include or not include the hydrogens:

removeHs=True

removeHs=False

If you would like to have RDKit manually add hydrogens, you can utilize the function Chem.AddHs().

**Warning - if you use this method, you must then run the AllChem.EmbedMolecule() function to optimize the 3D modeling again, otherwise the molecule will break. See below**

```python
rem2 = Chem.AddHs(rem)
drawit(rem2)
```

Instead, perform the following:

```python
rem2 = Chem.AddHs(rem)
AllChem.EmbedMolecule(rem2)
drawit(rem2)
```
